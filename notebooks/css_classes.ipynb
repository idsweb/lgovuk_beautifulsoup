# CSS Classes

Leave only allowed CSS classes

1. Get the main page content

```python
#!pip install beautifulsoup4

# Install the parsersc
#!pip install html5lib
#!pip install lxml

from bs4 import BeautifulSoup

with open("..\sourcefiles\Leeds youth justice service.html") as fp:
    noise_soup = BeautifulSoup(fp, 'html.parser')

# 
content = noise_soup.find_all('div', { 'id':'ctl00_PlaceHolderMain_ctl03__ControlWrapper_RichHtmlField'}) 
print(content[0])
```

Create a dictionary to hold the names of classes and their tags

```python
classes_to_keep = {}

# div tags
classes_to_keep['div'] = ['text-block','text-block']

for dirty_tag in content[0].select('[class]') :
    clean_tags = []
    for class_name in dirty_tag.attrs['class'] :
        if( dirty_tag.name in classes_to_keep):
            if(class_name in classes_to_keep[dirty_tag.name]) :
                clean_tags.append(class_name)
    if(len(clean_tags) > 0) :
        dirty_tag.attrs['class'] = clean_tags
    else :
        del(dirty_tag.attrs['class'])

print(content[0])
```
